# Acquisition des données

## Types de données

Sur le plan de l'acquisition des données, il distingue plusieurs types de données. Une première distinction utile est leur structure. On parle de données structurées comme les enregistrements d’une base de données relationnelle (dites SQL), les données semi-structurées d’une base de données non relationnelle (NoSQL) et les données non-structurées comme des textes libres, les traces, les commentaires dans un réseau social ou un forum,  ou de gros objets binaires comme des images, des photos, des sons ou des séquences vidéo qui sont le plus souvent stockés sous forme de fichiers dans des répertoires gérés par le système d’exploitation.

Typiquement les données structurées décrivent l’état d’objets à partir de leurs attributs ou caractéristiques. Par exemple, chacun des apprenants sera décrits par les attributs: nom, prénom, identifiant unique, adresse courriel, numéro de téléphone, niveau d’éducation, date de naissance, genre, adresse physique, code postal, etc. Les différents attributs utilisés pourront varier en fonction des besoins, de la culture et des politiques de protection de la vie privée.  Aussi certains attributs ont des formats normalisés comme les dates et les adresses de courriel.
Comme leur nom l’indique, les bases de données relationnelles exploitent les relations entre les données au travers un langage de requête structurée, le SQL qui permet d’organiser les informations en tables relationelles, de les interroger et les manipuler (modifier, ajouter, supprimer). 

Les bases de données non relationnelles, NoSQL, souvent simplement qualifiées d’outils de stockage de données, ont été créées autour de 2010 pour répondre aux besoins du traitement distribué de grandes quantités de données. Entre autres à cause de la difficulté inhérente au SQL de traiter efficacement certaines requêtes SQL comme la jointure sur des architectures distribuées. Le plus souvent les bases de données NoSQL gèrent des données structurées mais modifiées très souvent, faiblement structurées ou semi-structurées, ou encore structurées selon des modèles de données souples comme le JSON (JavaScript Object Notation) ou XML (eXtensible Markup Language).

## Types d'attributs

Sur le plan des types d’attributs, on distingue plusieurs types selon les valeurs qui les caractérisent. En gros, on distinguera entre les attributs discrets et les attributs continus. Une autre distinction utile sépare les attributs entre les attributs numériques avec des sous-types valeurs entières (valeurs discrètes) et valeurs réelles (valeurs continues), et les attributs catégoriels qui sont des symboles appartenant à un ensemble (valeurs discrètes). 

* **Valeurs continues**<br><br>
    * Par exemple, la température, le poids ou la taille d'une personne (des attributs continus) seront exprimés par un nombre en virgule flottante. 
    * Pour les besoins des algorithmes d’apprentissage automatique, les attributs numériques continus sont généralement convertis et normalisés en un nombre flottant entre 0.0 et 1.0
    * Pour les besoins des algorithmes d’apprentissage automatique, les attributs continus sont généralement convertis et normalisés en un nombre flottant entre 0 et 1.
<br><br>
* **Valeurs discrètes**<br><br>
    * Typiquement avec un ensemble précis et limité de cas possibles ou catégories (attributs catégoriels). Comme une valeur booléenne (vrai ou faux), le genre d'une personne (femme, homme), la couleur d'une voiture (noir, blanc, gris, bleu, vert, rouge, beige, ...), la réponse à une question à choix multiple. Même un texte peut être ramené à l'ensemble des mots qu'il contient.
    * Les différentes valeurs discrètes seront converties en une série d'attributs binaires (0,1) ou bits, à raison d'un bit par choix d'attribut. Note: Sauf pour le cas dégénéré à deux choix (i.e. attribut booléen) où un seul bit suffit.

## Examen des distributions des attributs numériques et des attributs catégoriels

### Boîte à moustaches
Une boîte à moustaches est une méthode d'affichage permettant de représenter graphiquement la distribution des données sur la base de cinq valeurs (le «minimum», le premier quartile (Q1), la médiane (ligne du milieu), le troisième quartile (Q3) et le «maximum»). Un cadre s'étend des valeurs du quartile Q1 au quartile Q3 des données, avec une ligne à la médiane (Q2). Les moustaches s’étendent à partir des bords de la boîte pour indiquer la plage des données.

La boîte à moustaches peut renseigner sur les valeurs aberrantes (outliers) et leurs valeurs. Il peut également vous indiquer si vos données sont symétriques, à quel point vos données sont groupées et si vos données sont asymétriques.

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from pandas.api.types import is_string_dtype, is_numeric_dtype 

NBR_MAX_ATTRIBUTS_AFFICHES = 50

# Créer une boîte à moustaches ou une listes de valeurs possibles
# pour chaque attribut ou colonne d'un tableau de données

def examen_distributions_attributs(un_dataframe):
    for attribut in un_dataframe.columns:
        # Afficher le type d'atribut
        print("-"*90)
        print("Attribut:",attribut,", Type:",un_dataframe.dtypes[attribut])
        # Si c'est un attribut catégoriel alors lister toutes les valeurs possibles
        if is_string_dtype(un_dataframe[attribut]):
            attributs_uniques = list(un_dataframe[attribut].unique())
            print("Nombre d'attributs_uniques:",len(attributs_uniques))
            if len(attributs_uniques) > NBR_MAX_ATTRIBUTS_AFFICHES:
                print("Affichage des " + str(NBR_MAX_ATTRIBUTS_AFFICHES) + " premières valeurs:")
                print(attributs_uniques[:NBR_MAX_ATTRIBUTS_AFFICHES])
            else:
                print(attributs_uniques)
                print("\nun_dataframe.groupby('"+attribut+"').size()\n",un_dataframe.groupby(attribut).size())
        if is_numeric_dtype(un_dataframe[attribut]):
            _ = un_dataframe.boxplot(attribut, figsize=(4, 6))
            plt.show()
            print()
            plt.close()
            _ = un_dataframe[attribut].plot(kind='hist',figsize=(4, 6))
            plt.show()
        print()
    print("-"*90)
    
examen_distributions_attributs(titanic_data_df)    
```

## Analyse des corrélations - fonction `.corr()`

Les étapes suivantes de l'analyse des données vont s’attacher à trouver des corrélations, à relever des tendances, identifier des relations, reconnaître des formes, des structures et des comportements et même faire des prédictions.

Par exemple examinons les différentes corrélations entre les attributs du jeu de données du Titanic.

En statistique, la corrélation entre variables dénote une dépendance. Le plus souvent, il s'agit de la corrélation linéaire ou coefficient de corrélation de Pearson toujours compris entre 0 et 1. Un coefficient nul n’implique pas indépendance, car d’autres types de corrélation sont possibles. Attention! Le fait que deux variables soient « fortement corrélées » ne montre pas qu'il y ait une relation de causalité.

```python
titanic_data_no_id_df.corr()
```

### Visualisation des corrélations
Maintenant rendons ces résultats plus visibles en ajoutant de la couleur...
Ici le rouge indique une corrélation positive, le blanc la neutralité et le bleu une corrélation négative. Évidemment, la diagonale est rouge vif!

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Une rustine un peu compliquée pour faire correspondre 0.0 à la couleur blanche
# https://stackoverflow.com/questions/20144529/shifted-colorbar-matplotlib/20146989#20146989
from matplotlib.colors import Normalize
class MidpointNormalize(Normalize):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))

norm = MidpointNormalize(midpoint=0)
    
alpha = ['Survived','Pclass','Age','SibSp','Parch','Fare']

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)
cax = ax.matshow(titanic_data_no_id_df.corr(), norm=norm, cmap='bwr', interpolation='nearest')
fig.colorbar(cax)

ax.set_xticklabels(['']+alpha)
ax.set_yticklabels(['']+alpha)

plt.show()
```

On voit que la survie est favorisée par la richesse et le fait d'avoir des parents ou des enfants qui nous accompagnent.

### Visualisation des tendances avec une matrice de dispersion

Une matrice de dispersion ou matrice de nuages de points (en anglais, scatter matrice) constitue une grille (ou matrice) de nuages de points permettant de visualiser les relations bivariées entre des combinaisons de variables numériques. Chaque nuage de points dans la matrice représente graphiquement la relation entre deux variables, ce qui permet d’explorer de nombreuses relations dans un diagramme.

Une matrice de nuages de points se compose d’une grille de nuages de points. Un histogramme montrant la distribution de chaque variable numérique se trouve sur la giagonale de la matrice. Un nuage de points est créé pour chaque combinaison de paires de variables.

Le résultat est un tableau carré avec dans la diagonale un histogramme pour chaque attribut et hors-diagonale des diagrammes de dispersion de l'attribut de la colonne par rapport à l'attribut de la ligne que l'on peut également colorer selon les différentes valeurs de l'étiquette cibles.

* Article de <a href="http://benalexkeen.com/scatter-matrices-using-pandas/">Ben Alex Keen</a> 

```python
# Retrait des données avec des valeurs manquantes
titanic_data_no_id_df = titanic_data_no_id_df.dropna()
etiquettes_cibles = titanic_data_no_id_df['Survived']
attributs = titanic_data_no_id_df.drop('Survived', axis=1)
```

```python
colMap={0:"red",1:"blue"}
colors_list=list(map(lambda x:colMap.get(x),etiquettes_cibles))
_ = pd.plotting.scatter_matrix(attributs, 
                           alpha=0.2, 
                           c=colors_list, 
                           figsize=[10,10])
plt.suptitle("Matrice de dispersion bivariée des attributs et colorés selon l'attribut 'Survived'", 
             fontsize=16)
print()
plt.show()
```

```python
colMap={0:"red",1:"blue"}
colors_list=list(map(lambda x:colMap.get(x),etiquettes_cibles))
_ = pd.plotting.scatter_matrix(attributs, 
                           alpha=0.2, 
                           c=colors_list, 
                           figsize=[10,10],
                           diagonal='kde')
plt.suptitle("Matrice de dispersion bivariée des attributs colorés selon l'attribut 'Survived'", 
             fontsize=16)
print()
plt.show()
```
